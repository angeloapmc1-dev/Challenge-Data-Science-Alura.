### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python

facturacion_tienda1 = tienda.Precio.sum()
facturacion_tienda2 = tienda2.Precio.sum()
facturacion_tienda3 = tienda3.Precio.sum()
facturacion_tienda4 = tienda4.Precio.sum()


print(f"Ventas Totales Tienda 1: ${facturacion_tienda1}")
print(f"Ventas Totales Tienda 2: ${facturacion_tienda2}")
print(f"Ventas Totales Tienda 3: ${facturacion_tienda3}")
print(f"Ventas Totales Tienda 4: ${facturacion_tienda4}")
```

# 2. Ventas por categoría

```python
print("--- Categorías más vendidas en la Tienda 4 ---")
print(tienda4['Categoría del Producto'].value_counts())
```

# 3. Calificación promedio de la tienda


```python

promedio_t1 = tienda.Calificación.mean()
promedio_t2 = tienda2.Calificación.mean()
promedio_t3 = tienda3.Calificación.mean()
promedio_t4 = tienda4.Calificación.mean()


print(f"Calificación Promedio Tienda 1: {promedio_t1:.2f}")
print(f"Calificación Promedio Tienda 2: {promedio_t2:.2f}")
print(f"Calificación Promedio Tienda 3: {promedio_t3:.2f}")
print(f"Calificación Promedio Tienda 4: {promedio_t4:.2f}")
```

# 4. Productos más y menos vendidos

```python
print("--- Los 5 productos MAS vendidos (Tienda 4) ---")
print(tienda4['Producto'].value_counts().head(5))

print("\n--- Los 5 productos MENOS vendidos (Tienda 4) ---")
print(tienda4['Producto'].value_counts().tail(5))
```

# 5. Envío promedio por tienda

```python
envio_t1 = tienda['Costo de envío'].mean()
envio_t2 = tienda2['Costo de envío'].mean()
envio_t3 = tienda3['Costo de envío'].mean()
envio_t4 = tienda4['Costo de envío'].mean()

print(f"Costo de Envío Promedio Tienda 1: {envio_t1:.2f}")
print(f"Costo de Envío Promedio Tienda 2: {envio_t2:.2f}")
print(f"Costo de Envío Promedio Tienda 3: {envio_t3:.2f}")
print(f"Costo de Envío Promedio Tienda 4: {envio_t4:.2f}")
```

```python
import matplotlib.pyplot as plt

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']
ventas = [facturacion_tienda1, facturacion_tienda2, facturacion_tienda3, facturacion_tienda4]

plt.figure(figsize=(10, 6))
plt.bar(tiendas, ventas, color=['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728'])

plt.title('Ventas Totales por Tienda')
plt.ylabel('Monto en CLP')
plt.show()
```
