```python
import pandas as pd
import requests

url = "https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/main/TelecomX_Data.json"

response = requests.get(url)
datos_json = response.json()

df = pd.json_normalize(datos_json)

df.head()
```

```python
df.info()
```

```python
df['account.Charges.Total'] = pd.to_numeric(df['account.Charges.Total'], errors='coerce')

print(df['account.Charges.Total'].isnull().sum())

df.info()
```

```python
df = df[df['Churn'] != '']

df.reset_index(drop=True, inplace=True)

df.info()
```

```python
df.dropna(subset=['account.Charges.Total'], inplace=True)

df.reset_index(drop=True, inplace=True)

df.info()
```

```python
diccionario = {'Yes': 1, 'No': 0}

df['Churn'] = df['Churn'].map(diccionario)

df[['customerID', 'Churn']].head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 6))
sns.countplot(x='Churn', data=df)
plt.title('Distribución de Evasión de Clientes (Churn)')
plt.show()

print(df['Churn'].value_counts(normalize=True) * 100)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(x='account.Contract', hue='Churn', data=df)
plt.title('Evasión por Tipo de Contrato')
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x='account.PaymentMethod', hue='Churn', data=df)
plt.title('Evasión por Método de Pago')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.kdeplot(df[df['Churn'] == 0]['account.Charges.Monthly'], label='Se quedan', fill=True)
sns.kdeplot(df[df['Churn'] == 1]['account.Charges.Monthly'], label='Se van', fill=True)
plt.title('Distribución de Cargos Mensuales por Churn')
plt.xlabel('Cargos Mensuales')
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Matriz de Correlación')
plt.show()
```

Conclusiones del Análisis de Evasión (Telecom X):

Se identificó una tasa de Churn del 26.58%, lo que representa una alerta para la empresa.

Los clientes con contrato "Month-to-month" y aquellos que utilizan "Electronic check" son los más propensos a abandonar el servicio.

Existe una relación directa entre los cargos mensuales altos (rango 70-110) y la fuga de clientes, confirmada por la densidad de distribución y la matriz de correlación.
